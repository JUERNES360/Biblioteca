```python
#Realizar un gráfico personalizado (aplicar un estilo atractivo)
#donde se pueda analizar las ventas de los 8 productos (Producto A, Producto B y Producto C, ingresar los que faltan)
#en tres ciudades (Bogotá, Cali, Bucaramanga).

import matplotlib.pyplot as plt

#Paso 2: Datos de Ventas por ciudad
Productos = ['Pr A','Pr B','Pr C','Pr D','Pr E','Pr F','Pr G','Pr H',] #listas con corchete
Bogota = [150, 100, 90, 234, 321, 455, 222, 111] # changed to 3 elements to match productos
Cali = [200, 180, 150, 222, 333, 444, 555, 666] # changed to J elements to match productos
Bucaramanga = [200, 220, 300, 400, 432, 234, 123, 234 ] # changed to 3 elements to match productos

#Paso 3: Crear graficos de barras apilados
plt.bar(Productos, Bogota, label='Bogotá', color='blue')
plt.bar(Productos, Cali, label='Cali', color='green', bottom=Bogota)
plt.bar(Productos, Bucaramanga, bottom=[i+j for i,j in zip(Bogota, Cali)], label='Bucaramanga', color='red')

#Paso 4: Añadir etiquetas, titulo y leyenda
plt.xlabel('Productos')
plt.ylabel('Ventas (Unidades)')
plt.title('comparacion de ventas por producto y ciudad')
plt.legend()
```

```python
# Realizar un gráfico personalizado (aplicar un estilo atractivo) para visualizar
#la evolución de las ventas de tres productos (Papa, Cebolla, Tomate)
#a lo largo de 8 meses en tres ciudades diferentes.

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

#Paso 2: Crear el DataFrame con las ventas por productos y ciudad
datos = pd.DataFrame({
'Mes' : ['Enero','Febrero','Marzo','Abril','Mayo','Junio','Julio','Agosto'],
'Papa (Cali)' : [100, 150, 200, 250, 300, 400, 300, 150],
'Cebolla (Bogota)' : [90, 120, 160, 200, 240, 100, 200, 150],
'Tomate (Cartagena)': [80, 110, 150, 190, 230, 330, 350, 240]
})

#Paso 3: Configurar el estilo de seaborn
sns.set_theme(style="whitegrid")


plt.plot(datos['Mes'], datos['Papa (Cali)'], marker= 'o', color='blue', label='Papa (Cali)')
plt.plot(datos['Mes'], datos['Cebolla (Bogota)'], marker= 's', color='green', label='Cebolla(Bogota)')
plt.plot(datos['Mes'], datos['Tomate (Cartagena)'], marker= '^', color='red', label='Tomate (Cartagena)')

#Paso 4: Personalizar el grafico con titulos y etiquetas
plt.title('Evolucion de ventas por producto y ciudad', fontsize=14)
plt.xlabel('Mes', fontsize=12)
plt.ylabel('Ventas (Unidades)', fontsize=12)

#Añadir leyenda
plt.legend()

#Mostrar el grafico
plt.show()


```
